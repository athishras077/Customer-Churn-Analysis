```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/telco_clean.csv")
```

```python
# Section 1 – Univariate Analysis : (Analyze one variable at a time)
# Customer Churn Distribution

ax = sns.countplot(x="Churn Label", data=df, width=0.3)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Customer Churn Distribution")
plt.show()
```

```python
# Gender Distribution
ax = sns.countplot(x="Gender", data=df, width = 0.3)

for container in ax.containers:
    ax.bar_label(container)
    
plt.title("Gender Distribution")
plt.show()
```

```python
# Contract Type

ax = sns.countplot(x="Contract", data=df, width = 0.3)

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=20)
plt.show()
```

```python
# Internet Service Type

ax = sns.countplot(x="Internet Type", data=df, width = 0.3)

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=20)
plt.show()
```

```python
# Payment Method
ax = sns.countplot(x="Payment Method", data=df, width = 0.3)

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=20)
plt.show()
```

```python
# Monthly Charges Distribution

sns.histplot(df["Monthly Charge"], bins=30)
plt.show()
```

```python
# Tenure Distribution
sns.histplot(df["Tenure in Months"], bins=30)
plt.show()
```

```python
# Section 2 – Bivariate Analysis (Compare two variables)
# Purpose: Compare variables with churn.

# Churn by Gender
ax = sns.countplot(x="Gender", hue="Churn Label", data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Churn by Gender")
plt.show()
```

```python
# Churn by Senior Citizen
ax = sns.countplot(x="Senior Citizen", hue="Churn Label", data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Churn by Senior Citizen")
plt.show()
```

```python
# Churn by Contract
ax = sns.countplot(x="Contract", hue="Churn Label", data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Churn by Contract")
plt.show()
```

```python
# Churn by Internet Type
ax = sns.countplot(x="Internet Type", hue="Churn Label", data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Churn by Internet Type")
plt.show()
```

```python
# Churn by Payment Method
ax = sns.countplot(x="Payment Method", hue="Churn Label", data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Churn by Payment Method")
plt.show()
```

```python
# Monthly Charges vs Churn

sns.boxplot(x="Churn Label", y="Monthly Charge", data=df)
plt.show()
```

```python
# Tenure vs Churn

sns.boxplot(x="Churn Label", y="Tenure in Months", data=df)
plt.show()
```

```python
# Satisfaction Score vs Churn

sns.boxplot(data = df, x = "Churn Label", y = "Satisfaction Score")
plt.show()
```

```python
# Section 3 – Multivariate Analysis (Analyze multiple variables together)

# Contract vs Internet Type vs Churn

pd.crosstab(
    [df["Contract"], df["Internet Type"]], df["Churn Label"] )
```

```python
# Customer Status by Contract
pd.crosstab(df["Contract"], df["Customer Status"])
```

```python
# Monthly Charge by Contract

sns.boxplot(data = df, x = "Contract", y = "Monthly Charge")
plt.show()
```

```python
# Revenue by Customer Status

sns.boxplot(data = df, x ="Customer Status", y = "Total Revenue")
plt.show()
```

```python
# Section 4 – Correlation Analysis (Understand numerical relationships)

numeric_df = df.select_dtypes(include=["number"])

corr = numeric_df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.show()
```
